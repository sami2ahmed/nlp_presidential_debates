```python
# load dataset from https://www.kaggle.com/kinguistics/2016-us-presidential-primary-debates/version/1

# dataset contains transcripts of every Democratic and Republican debate held during the 2016 primary season 

debate_dem = pd.read_csv('primary_debates_cleaned.csv')
```

# Prepare data for Mongo 

```python
# quick clean and preview of df

debate_dem = debate_dem.drop(['Line'], axis=1)
debate_dem.tail(5)
```

```python
# convert df to .json 

document_dicts = []
for row in debate_dem.iterrows():
    document_dicts.append(row[1].to_dict())
```

```python
# preview .json make sure it looks good/ready for mongo 

document_dicts[:3]
```

```python
# pickle dictionary for later use, data is used again in 1_LSA_text_cleaning notebook 
 
with open('twenty16_debate_corpus.pkl', 'wb') as f:
    pickle.dump(document_dicts, f)
```

# Load data into Mongo 

```python
# get the .json into mongo as a collection (same as table in SQL)

db.twenty16_debate.insert_many(document_dicts)
```

```python
# confirm collection is in mongo 

db.collection_names()
```

```python
# print everything in the collection, ensure document format is good 

for doc in db.twenty16_debate.find({}):
    print(doc)
```
